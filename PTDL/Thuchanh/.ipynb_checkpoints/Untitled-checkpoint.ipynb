```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

lst = ['red', 'red', 'yellow', 'green', 'yellow']

data = pd.DataFrame(lst,columns = ['color'])
display(data)
```

```python
encoder = OneHotEncoder(sparse = False) #Tạo object encoder
onehot = encoder.fit_transform(data)
print(onehot)
```

```python
print(pd.get_dummies(data)) #Dùng hàm get_dummies() của pandas
```

```python
from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()
label = encoder1.fit_transform(data)
print(label)
```

```python
import numpy as np
conditions = [data_1['color'] == 'green',
              data_1['color'] == 'orange',
              data_1['color'] == 'pink',
              data_1['color'] == 'red',
              data_1['color'] == 'yellow']
choices = [1,2,3,4,5]
data_1['Encoder'] = np.select(conditions,choices,default = 'NaN')
print(data_1)
```

```python
from sklearn.preprocessing import OrdinalEncoder
list3 = ['excellent', 'very_good' 'good', 'normal', 'bad']
dat3 = pd.DataFrame(list3, columns = ['rate'])
encoder3 = OrdinalEncoder()
oe = encoder3.fit_transform(dat3)
print(oe)

# list_.reverse()
encoder3.categories = [np.array(list3)]
oe2 = encoder3.fit_transform(dat3)
print(oe2)
```
