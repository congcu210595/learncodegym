```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler

df = pd.read_excel("house_price_dống-da.xlsx")
df.info()
```

```python
df.head()
```

```python
#Phát hiện các dòng, cột chứa dữ liệu khuyết thiếu
df.isna()
```

```python
#Xóa bỏ hết tất cả những dòng dữ liệu không có thông tin về giá
df1 = df.dropna(subset = ['price'])
```

```python
print(df1['house_direction'].mode())
print(df1['balcony_direction'].mode())
print(df1['toilet'].mode())
print(df1['bedroom'].mode())
print(df1['floor'].mode())
```

```python
#Thực hiện xử lý giá trị khuyết thiếu: Thay thế giá trị khuyết thiếu của land_certificate bằng =”không có thông tin”, house_direction, balcony_direction, toilet, bedroom, Floor  bằng giá trị có tần số xuất hiện lớn nhất của các thuộc tính đó

values = {'land_certificate': 'không có thông tin',
          'house_direction': 'Đông-Nam',
          'balcony_direction': 'Đông-Nam',
          'toilet': 4.0,
          'bedroom': 4.0,
          'floor': 5.0}
df2 = df1.fillna(value = values)

print(df2.head())
```

```python
#Lọc thông tin những bất động sản ở trong ngõ thành bộ dữ liệu nhà ngõ
df3 = df2[(df2['type_of_land'] == df2['type_of_land'][0]) | (df2['type_of_land'] == 'Bán nhà riêng')]
df3.head()
```

```python
#Tính toán giá/m2  ( đơn vị triệu/m2) với loại hình nhà ngõ
df3.drop(df3[df3['area'] == 0].index, inplace=True)
df3 = df3.dropna(subset = ['area'])
df3['Giá/m2'] = df3['price'] / df3['area']
df3.head()
```

```python
df3 = df3[['area','Giá/m2']]
df3.boxplot()
```

```python
#Phát hiện giá trị ngoại lai của các thuộc tính: diện tích, giá/m2 bằng phương pháp IQR

Q1 = df3.quantile(0.25)
Q3 = df3.quantile(0.75)
IQR = Q3-Q1
df4 = df3[~((df3 < (Q1 - 1.5 * IQR)) | (df3 > (Q3 + 1.5 * IQR))).any(axis=1)]
df4.boxplot()
```

```python
#Chuẩn hóa dữ liệu bằng các phương pháp: min-max scaling

scaler = MinMaxScaler()

mms = scaler.fit_transform(pd.DataFrame(df4['Giá/m2']))
sns.kdeplot(data = mms)
```

```python
# chuẩn hóa với robust scaling
scaler = RobustScaler()

rbs = scaler.fit_transform(pd.DataFrame(df4['Giá/m2']))
sns.kdeplot(data = rbs)
```

```python
# chuẩn hóa với standard scaling
scaler = StandardScaler()

sc = scaler.fit_transform(pd.DataFrame(df3['Giá/m2']))
sns.kdeplot(data = sc)
```

```python
print('Có thể nhận thấy ở 2 phương pháp chuẩn hóa min-max scaling và robust scaling, phân bố dữ liệu gần như được giữ nguyên, tuy nhiên ở phương pháp standard scaling có sự thay đổi,khoảng giá trị giữ liệu cũng đã thay đổi khác nhau')
```
