```python
import pygame, sys, random
from pygame.locals import *

sc = pygame.display.set_mode((400,600))
pygame.display.set_caption('RACING')
FPS = 60
fpsClock = pygame.time.Clock()

bg_img = pygame.image.load('background.png')
bg_s = 1.5

car_w = 40
car_h = 60
car_s = 3
car_img = pygame.image.load('car.png')

ob_s = 2
ob_img = pygame.image.load('obstacles.png')

pygame.init()

y_bg = 0
def draw_bg():
    global y_bg
    sc.blit(bg_img,(0,y_bg))
    sc.blit(bg_img,(0,y_bg - bg_img.get_height()))
    y_bg += bg_s
    if y_bg > bg_img.get_height():
        y_bg -= bg_img.get_height()

x_car = (400-car_w)/2
y_car = (600-car_h)/2
def draw_car(Left,Right,Up,Down):
    global x_car
    global y_car
    sc.blit(car_img,(x_car,y_car))
    if Left and x_car > 80:
        x_car -= car_s
    if Right and x_car + car_w < 320:
        x_car += car_s
    if Up and y_car > 0:
        y_car -= car_s
    if Down and y_car + car_h < 600:
        y_car += car_s
        
ls_ob = []
def cr_ob():
    for i in range(5):
        y_ob1 = -car_h - i* 200
        lane_ob1 = random.randint(0,3)
        ls_ob.append([lane_ob1,y_ob1])
def draw_ob():
    global ob_s
    for i in range(5):
        x_ob = int(80 + ls_ob[i][0]*60 + (60 - car_w)/2)
        sc.blit(ob_img,((x_ob,int(ls_ob[i][1]))))
    for i in range(5):
        ls_ob[i][1] += ob_s
    ob_s += 0.001
    if ls_ob[0][1] > 600:
        ls_ob.pop(0)
        y_ob = int(ls_ob[3][1] - 200)
        lane_ob = random.randint(0,3)
        ls_ob.append([lane_ob,y_ob])
        
score = 0
def draw_score():
    global score
    font = pygame.font.SysFont('consolas', 30)
    scoreSuface = font.render('Score: '+str(int(score)), True, (0, 0, 0))
    sc.blit(scoreSuface, (10, 10))
    score += 0.015
        
def rectCollision(rect1, rect2):
    if rect1[0] <= rect2[0]+rect2[2] and rect2[0] <= rect1[0]+rect1[2] and rect1[1] <= rect2[1]+rect2[3] and rect2[1] <= rect1[1]+rect1[3]:
        return True
    return False

def isGameover():
    carRect = [x_car,y_car,car_w, car_h]
    for i in range(5):
        x_ob = 80 + ls_ob[i][0]*60 + (60-car_w)/2
        y_ob = ls_ob[i][1]
        obstaclesRect = [x_ob, y_ob, car_w, car_h]
        if rectCollision(carRect, obstaclesRect) == True:
            return True
    return False

def gameStart():
    
    font = pygame.font.SysFont('consolas', 60)
    headingSurface = font.render('RACING', True, (255, 0, 0))
    headingSize = headingSurface.get_size()

    font = pygame.font.SysFont('consolas', 20)
    commentSurface = font.render('Press "space" to play', True, (0, 0, 0))
    commentSize = commentSurface.get_size()
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == pygame.KEYUP:
                if event.key == K_SPACE:
                    return
        sc.blit(bg_img,(0,0))
        sc.blit(headingSurface, (int((400 - headingSize[0])/2), 100))
        sc.blit(commentSurface, (int((400 - commentSize[0])/2), 400))
        pygame.display.update()
        fpsClock.tick(FPS)

def gamePlay():
    m_Left = False
    m_Right = False
    m_Up = False
    m_Down = False
    cr_ob()
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == KEYDOWN:
                if event.key == K_LEFT:
                    m_Left = True
                if event.key == K_RIGHT:
                    m_Right = True
                if event.key == K_UP:
                    m_Up = True
                if event.key == K_DOWN:
                    m_Down = True
            if event.type == KEYUP:
                if event.key == K_LEFT:
                    m_Left = False
                if event.key == K_RIGHT:
                    m_Right = False
                if event.key == K_UP:
                    m_Up = False
                if event.key == K_DOWN:
                    m_Down = False

        draw_bg()
        draw_car(m_Left,m_Right,m_Up,m_Down)
        draw_ob()
        draw_score()
        if isGameover():
            return
        pygame.display.update()
        fpsClock.tick(FPS)
        
def gameOver():
    global ls_ob
    global x_car
    global y_car
    global score
    global ob_s    
    font = pygame.font.SysFont('consolas', 60)
    headingSurface = font.render('GAMEOVER', True, (255, 0, 0))
    headingSize = headingSurface.get_size()

    font = pygame.font.SysFont('consolas', 40)
    scoreSurface = font.render('SCORE: ' + str(int(score)), True, (0,0,255))
    scoreSize = scoreSurface.get_size()  
    
    font = pygame.font.SysFont('consolas', 20)
    commentSurface = font.render('Press "space" to replay', True, (0, 0, 0))
    commentSize = commentSurface.get_size()
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == pygame.KEYUP:
                if event.key == K_SPACE:
                    ob_s = 2
                    x_car = (400-car_w)/2
                    y_car = (600-car_h)/2
                    ls_ob = []
                    score = 0
                    return
        sc.blit(bg_img,(0,0))
        sc.blit(headingSurface, ((400 - headingSize[0])/2, 100))
        sc.blit(commentSurface, ((400 - commentSize[0])/2, 400))
        sc.blit(scoreSurface, ((400 - scoreSize[0])/2, 250))
        
        pygame.display.update()
        fpsClock.tick(FPS)

gameStart()
while True:
    gamePlay()
    gameOver()
```
