```python
import turtle
t = turtle.Turtle()
t.pensize(3)
t.hideturtle()

def hinhcn(a,b,c):
    t.fillcolor(c)
    t.begin_fill()
    for i in range(2):
        t.fd(a)
        t.rt(90)
        t.fd(b)
        t.rt(90)
    t.end_fill()
    
def hinhtamgiacvuongcan(a,b):
    t.fillcolor(b)
    t.begin_fill()
    t.rt(45)
    t.fd(a)
    t.rt(135)
    t.fd(a*2**0.5)
    t.rt(135)
    t.fd(a)
    t.rt(45)
    t.end_fill()
    
def hinhthangcan(a,b,c,d,e):
    t.fillcolor(e)
    t.begin_fill()
    t.fd(a/2)
    t.rt(180-c)
    t.fd(d)
    t.rt(c)
    t.fd(b)
    t.rt(c)
    t.fd(d)
    t.rt(180-c)
    t.fd(a/2)
    t.end_fill()

t.goto(-150,0)
hinhcn(300,130,"yellow")

t.goto(0,0)
t.lt(90)
hinhcn(85,20,"brown")
t.rt(90)

t.pu()
t.goto(-50,100)
t.pd()
hinhtamgiacvuongcan(141.42,"orange")


t.penup()
t.goto(50,0)
t.bk(30)
t.lt(90)
t.fd(100)
t.pendown()
t.fd(40)
t.lt(90)
t.penup()
t.fd(15)
t.lt(90)
t.fd(10)
t.pendown()
t.fd(25)

t.penup()
t.goto(35,-40)
t.pendown()
hinhcn(90,70,"blue")

t.penup()
t.goto(320,120)
t.pendown()
t.lt(90)
hinhtamgiacvuongcan(60,"green")

t.penup()
t.goto(320,77.57)
t.pendown()
hinhtamgiacvuongcan(90,"green")

t.penup()
t.goto(320,13.93)
t.pendown()
hinhtamgiacvuongcan(120,"green")

t.penup()
t.goto(320,-70.92)
t.pendown()
hinhthangcan(30,50,95,130,"brown")

t.penup()
t.goto(-200,-40)
t.pendown()
hinhtamgiacvuongcan(30,"green")

t.penup()
t.goto(-200,-61.2)
t.pendown()
hinhtamgiacvuongcan(45,"green")

t.penup()
t.goto(-200,-93.02)
t.pendown()
hinhtamgiacvuongcan(60,"green")

t.penup()
t.goto(-200,-135.45)
t.pendown()
hinhthangcan(15,25,95,65,"brown")
```
