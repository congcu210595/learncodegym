```python
import random
dem = 0
if random.randint(0,1) == 0:
    current_player = "human"
else:
    current_player = "computer"
while True:
    print("tổng hiện tại =",dem)
    if current_player == "human":
        player_choice = input("Người chơi chọn số: ")
        while player_choice != "1" and player_choice != "2" and player_choice != "3":
            player_choice = input("Người chơi chọn số: ")
        dem += int(player_choice)
        if dem >= 21:
            print("tổng hiện tại =",dem,"Người chơi đã thua")
            play_again = input("Play again? : ")
            if play_again[0] == "y":
                dem = 0
            else:
                break
        else:
            current_player = "computer"
    elif current_player == "computer":
        computer_choice = random.randint(1,3)
        print("Máy chọn số: ",computer_choice)
        dem += computer_choice
        if dem >= 21:
            print("tổng hiện tại =",dem,"Người chơi đã thắng")
            play_again = input("Play again? : ")
            if play_again[0] == "y":
                dem = 0
            else:
                break
        else:
            current_player = "human"
            
        
         
            
        
```
