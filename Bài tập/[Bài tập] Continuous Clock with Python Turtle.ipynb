```python
import turtle
import datetime

t = turtle.Turtle()
sc = turtle.Screen()
t.pencolor("black")
t.pensize(3)
t.speed(0)
sc.tracer(0,0)
t.ht()

def ve_dong_ho(ms,s,m,h):
    t.clear()
    
    t.circle(180)
    t.pu()
    t.goto(0,180)
    for i in range(60):
        if i % 5 == 0:
            t.fd(135)
            t.pd()
            t.fd(25)
            t.pu()
        else:
            t.fd(150)
            t.pd()
            t.fd(10)
            t.pu()        
        t.bk(160)
        t.rt(6)
        
    t.pu()
    t.goto(0,180)
    t.pd()
    t.pensize(1)
    t.lt(90)
    t.rt(6*s + 6*ms/1000000)
    t.fd(120)
    t.pu()
    t.goto(0,180)
    t.lt(6*s + 6*ms/1000000)
    
    t.pd()
    t.pensize(3)
    t.rt(6*m + 6*s/60)
    t.fd(100)
    t.pu()
    t.goto(0,180)
    t.lt(6*m + 6*s/60)
    
    t.pd()
    t.pensize(5)
    t.rt(30*(h%12) + 30*m/60 + 30*s/3600)
    t.fd(80)
    t.pu()
    t.goto(0,180)
    t.lt(30*(h%12) + 30*m/60 + 30*s/3600)
    t.rt(90)
    
def update():
    d = datetime.datetime.now()
    ms = d.microsecond
    s = d.second
    m = d.minute
    h = d.hour 
    ve_dong_ho(ms,s,m,h)
    sc.update()
    sc.ontimer(update,10)
    

sc.update()
update()
sc.mainloop()
```
