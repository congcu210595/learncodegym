```python
from tkinter.ttk import *
from tkinter import *

sc = Tk()
sc.title("Calulator")
sc.config(bg = 'black')

height = 2
width = 10
relief = "ridge"
font = ("Arial",14,'bold')
bg = 'gray'
fg = 'yellow'

noidung_et = StringVar()

Entry(sc, text="",font = ("Arial",25),textvariable = noidung_et).grid(column=0, row=0, sticky = "WE",columnspan = 4)

def nhan_1():
    noidung_et.set(noidung_et.get() + "1")
    
def nhan_2():
    noidung_et.set(noidung_et.get() + "2")
    
def nhan_3():
    noidung_et.set(noidung_et.get() + "3")

def nhan_4():
    noidung_et.set(noidung_et.get() + "4")
    
def nhan_5():
    noidung_et.set(noidung_et.get() + "5")
    
def nhan_6():
    noidung_et.set(noidung_et.get() + "6")
    
def nhan_7():
    noidung_et.set(noidung_et.get() + "7")
    
def nhan_8():
    noidung_et.set(noidung_et.get() + "8")
    
def nhan_9():
    noidung_et.set(noidung_et.get() + "9")

def nhan_0():
    noidung_et.set(noidung_et.get() + "0")

def nhan_cong():
    noidung_et.set(noidung_et.get() + "+")
    
def nhan_tru():
    noidung_et.set(noidung_et.get() + "-")
    
def nhan_nhan():
    noidung_et.set(noidung_et.get() + "*")
    
def nhan_chia():
    noidung_et.set(noidung_et.get() + "/")
    
def nhan_mongoac():
    noidung_et.set(noidung_et.get() + "(")
    
def nhan_dongngoac():
    noidung_et.set(noidung_et.get() + ")")

def nhan_cham():
    noidung_et.set(noidung_et.get() + ".")
    
def nhan_mr(): 
    noidung_et.set("")
    
def nhan_del(): 
    noidung_et.set(noidung_et.get()[:-1])
    
def nhan_bang():
    ketqua = eval(noidung_et.get())
    noidung_et.set(ketqua)

Button(sc, text="MR", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_mr).grid(column=0, row=1)
Button(sc, text="DEL", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_del).grid(column=1, row=1)
Button(sc, text="(", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_mongoac).grid(column=2, row=1)
Button(sc, text=")", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_dongngoac).grid(column=3, row=1)

Button(sc, text="1", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_1).grid(column=0, row=2)
Button(sc, text="2", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_2).grid(column=1, row=2)
Button(sc, text="3", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_3).grid(column=2, row=2)
Button(sc, text="+", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_cong).grid(column=3, row=2)

Button(sc, text="4", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_4).grid(column=0, row=3)
Button(sc, text="5", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_5).grid(column=1, row=3)
Button(sc, text="6", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_6).grid(column=2, row=3)
Button(sc, text="-", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_tru).grid(column=3, row=3)

Button(sc, text="7", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_7).grid(column=0, row=4)
Button(sc, text="8", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_8).grid(column=1, row=4)
Button(sc, text="9", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_9).grid(column=2, row=4)
Button(sc, text="*", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_nhan).grid(column=3, row=4)

Button(sc, text=".", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_cham).grid(column=0, row=5)
Button(sc, text="0", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_0).grid(column=1, row=5)
Button(sc, text="=", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_bang).grid(column=2, row=5)
Button(sc, text="/", height=height, width=width,relief = relief,font = font,bg = bg,fg = fg, command = nhan_chia).grid(column=3, row=5)

sc.mainloop()
```
