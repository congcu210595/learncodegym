```python
import turtle
t = turtle.Turtle()
t.pencolor("blue")
t.pensize(3)

def hinhcn(a,b,c):
    t.fillcolor(c)
    t.begin_fill()
    for i in range(2):
        t.fd(a)
        t.rt(90)
        t.fd(b)
        t.rt(90)
    t.end_fill()


hinhcn(200,200,"grey")
t.lt(90)
t.pu()
t.goto(60,-200)
t.pd()
hinhcn(100,40,"green")
t.pu()
t.goto(100,-200)
t.pd()
hinhcn(100,40,"green")

t.pu()
t.goto(0,0)
t.pd()
t.lt(90)
hinhcn(150,200,"red")
t.lt(90)
hinhcn(200,150,"red")
t.rt(90)
t.fd(150)
hinhcn(150,200,"red")
t.lt(90)
hinhcn(200,150,"red")

t.lt(90)
t.pu()
t.goto(-120,140)
t.pd()
hinhcn(40,70,"white")

t.pu()
t.goto(-250,140)
t.pd()
hinhcn(40,70,"white")

t.pu()
t.goto(-250,-60)
t.pd()
hinhcn(40,70,"white")

t.pu()
t.goto(-120,-60)
t.pd()
hinhcn(40,70,"white")
```
